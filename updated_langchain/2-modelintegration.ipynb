### Models integration with google gemini,groq and openAI

```python
import google.generativeai as genai
import os

genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))

for model in genai.list_models():
    print(model.name)
```

```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)

models = client.models.list()

for model in models.data:
    print(model.id)
```

```python
from groq import Groq
import os

client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)

models = client.models.list()

for model in models.data:
    print(model.id)
```

### Streaming and Batch

Streaming 
Most models can stream their output content while being generated.By displaying output progessively streaming significantly improves user experience particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced. We can use loops to process each chunk in real time

```python
model.invoke("write me a 200 word essay on the history of AI")
```

```python
for chunk in model.stream("write me a 200 word essay on the history of AI"):
    print(chunk.text,end="|",flush=True)
```

Batch
Collection of independent requests to a model ...as processing can be done in parallel to improve the performance and reduce costs

```python
responses=model.batch([
    "why are you dumb?",
    "write me a haiku about the ocean",
    "what is the meaning of life?"
    
])
for response in responses:
    print(response.text)
```
