# MESSAGES
Messages are the fundamental unit of context for models in langchain. they represent the input and the output of models carrying both the content and the metadata needed to represent the state of a conversation when interacting with a LLM. Measages are the obejcts that contain:

1) Role - identifies the message type(eg system, user)
2) Content - Represents the actual content of the message (like text,imag,documents)
3) Metadata - optional field such as response info, message IDs and token usage

Langchain provides a standard message type that works across all model providers ensuring consistent behaviour regardless of the model being called.

```python
import os
from dotenv import load_dotenv
load_dotenv()

from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
model.invoke("how are you? ")
```

### Text Prompts
Text prompts are strings- ideal for straightforward generation tasks where we dont need to retain conversation history

```python
model.invoke("what is a tree?")
```

Use text prompts when:
we have a single,standalone request
we dont need conversation history
we want minimal code complexity

### Message Prompts
Alternatvely we can pass in a list of messages to the model providing a list of message objects.

Messsage types:
1) System message- tells the model how to behave and provide context for interactions
2) Human message - represents user input and interactions with the model
3) AI meassage- Responses generated by the model , including text content, tool calls and metadata
4) Tool message - Represents the outputs of the tool calls


### System Message
A system Message represent an initial set of instructions that primes the model's behaviour. We can use a system meassage to set the tone and define model's role and establish guidelines for responses.

### Human Message
 A human message represent user input and interactions. they contain text,images,audio,files or any other multimodal content.

### AI Message
An AI message represents the output of a model invocation. they can include multimodal data, tool calls and provider specific metadata which can be later accessed by us.

### Tool Message
For models that support tool calling , AI messages can contain tool calls. Tool messages are used to paas the result of single tool execution back to the model.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
messages=[
    SystemMessage(content="You are an expert writer."),
    HumanMessage(content="Write a haiku about love")
]
response=model.invoke(messages)
response.content
```

```python
##Detailed info to LLM through system messages
from langchain.messages import SystemMessage, HumanMessage, AIMessage

system_msg=SystemMessage(""" You are a senior python developer with expertise in web frameworks.Always provide code examples and explain your reasoning. Be concise but thorough in your explanations""")

messages=[
    system_msg,
    HumanMessage(content="How do i create REST API?")
]
response=model.invoke(messages)
response.content
```

```python
from langchain.messages import HumanMessage
human_msg=HumanMessage(
    content="I am very angry with you",
    name="subhav",
    id="1234"
)
response=model.invoke([human_msg])
response.content
```

```python
ai_msg=AIMessage("I'd be happy to help you with that question")
messages=[
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="Can you help me"),
    ai_msg,
    HumanMessage(content="Great! What is 2+2?", name="subhav", id="1234")
]
response=model.invoke(messages)
response.content
```

```python
response.usage_metadata
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage
ai_message=AIMessage(
    content=[],
    tool_calls=[
        {
            "name":"get_weather",
            "args":{
                "location":"New York"
            },
            "id":"1234"
        }
    ]
)
weather_result="Sunny, 25 degrees Celsius"
tool_message=ToolMessage(
    content=weather_result,
    tool_call_id="1234"
)

messages=[
    HumanMessage(content="What's the weather like in New York?"),
    ai_message,
    tool_message
    
]
response=model.invoke(messages)
```
